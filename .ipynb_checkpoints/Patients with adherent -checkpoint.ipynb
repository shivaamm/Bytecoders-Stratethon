```python
import pandas as pd
import numpy as np
```

```python
patients = pd.read_csv('./Case Files/sample_date_csv/patients.csv')
print('Total shape : ',patients.shape)
patients.info()
```

```python
# Droping unnecessary columns
patients.drop(['DEATHDATE','DRIVERS','PASSPORT','PREFIX','SUFFIX','MAIDEN','MARITAL','COUNTY','ZIP','LAT','LON'],axis = 1,inplace = True)
patients.head()
```

```python
isadherent = pd.read_csv('isadherent.csv')
isadherent.rename(columns = {'PATIENT':'Id'}, inplace = True)
isadherent.head()
```

```python
df = (pd.merge(patients, isadherent, on='Id'))
df.head()
```

```python
df.isnull().sum()
```

```python
df.to_csv('../myCSV/patients_with_isadherence.csv',index=False)
```
